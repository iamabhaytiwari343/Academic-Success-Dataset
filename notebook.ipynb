```python
train['Total curricular units enrolled'] = train['Curricular units 1st sem (enrolled)'] + train['Curricular units 2nd sem (enrolled)']
train['Total curricular units evaluated'] = train['Curricular units 1st sem (evaluations)'] + train['Curricular units 2nd sem (evaluations)']
train['Total curricular units approved'] = train['Curricular units 1st sem (approved)'] + train['Curricular units 2nd sem (approved)']
train['Total curricular units without evaluations'] = train['Curricular units 1st sem (without evaluations)'] + train['Curricular units 2nd sem (without evaluations)']
train['Overall average grade'] = (train['Curricular units 1st sem (grade)'] + train['Curricular units 2nd sem (grade)']) / 2
# train['Approval ratio 1st sem'] = train['Curricular units 1st sem (approved)'] / train['Curricular units 1st sem (enrolled)']
# train['Approval ratio 2nd sem'] = train['Curricular units 2nd sem (approved)'] / train['Curricular units 2nd sem (enrolled)']
# train['Evaluation ratio 1st sem'] = train['Curricular units 1st sem (evaluations)'] / train['Curricular units 1st sem (enrolled)']
# train['Evaluation ratio 2nd sem'] = train['Curricular units 2nd sem (evaluations)'] / train['Curricular units 2nd sem (enrolled)']
train['Is financially stable'] = train['Tuition fees up to date'].apply(lambda x: 1 if x == 'Yes' else 0)
train['Is scholarship holder'] = train['Scholarship holder'].apply(lambda x: 1 if x == 'Yes' else 0)
train['Is international student'] = train['International'].apply(lambda x: 1 if x == 'Yes' else 0)
train['Economic difficulty'] = (train['Unemployment rate'] + train['Inflation rate']) / train['GDP']

test['Total curricular units enrolled'] = test['Curricular units 1st sem (enrolled)'] + test['Curricular units 2nd sem (enrolled)']
test['Total curricular units evaluated'] = test['Curricular units 1st sem (evaluations)'] + test['Curricular units 2nd sem (evaluations)']
test['Total curricular units approved'] = test['Curricular units 1st sem (approved)'] + test['Curricular units 2nd sem (approved)']
test['Total curricular units without evaluations'] = test['Curricular units 1st sem (without evaluations)'] + test['Curricular units 2nd sem (without evaluations)']
test['Overall average grade'] = (test['Curricular units 1st sem (grade)'] + test['Curricular units 2nd sem (grade)']) / 2
# test['Approval ratio 1st sem'] = test['Curricular units 1st sem (approved)'] / test['Curricular units 1st sem (enrolled)']
# test['Approval ratio 2nd sem'] = test['Curricular units 2nd sem (approved)'] / test['Curricular units 2nd sem (enrolled)']
# test['Evaluation ratio 1st sem'] = test['Curricular units 1st sem (evaluations)'] / test['Curricular units 1st sem (enrolled)']
# test['Evaluation ratio 2nd sem'] = test['Curricular units 2nd sem (evaluations)'] / test['Curricular units 2nd sem (enrolled)']
test['Is financially stable'] = test['Tuition fees up to date'].apply(lambda x: 1 if x == 'Yes' else 0)
test['Is scholarship holder'] = test['Scholarship holder'].apply(lambda x: 1 if x == 'Yes' else 0)
test['Is international student'] = test['International'].apply(lambda x: 1 if x == 'Yes' else 0)
test['Economic difficulty'] = (test['Unemployment rate'] + test['Inflation rate']) / test['GDP']   
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
train['Target'] = encoder.fit_transform(train['Target'])
```

```python
ytrain=train['Target']
Xtrain=train.drop(['Target','id'],axis=1)
testid=test['id']
Xtest=test.drop('id',axis=1)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import HistGradientBoostingClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier

# Define the base classifiers
lgbm = LGBMClassifier()
catboost = CatBoostClassifier(verbose=0)
xgboost = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
knn = KNeighborsClassifier(algorithm='auto', n_neighbors=7, p=1, weights='uniform')
hist_gbc = HistGradientBoostingClassifier()

# Define the voting classifier
voting_model = VotingClassifier(
    estimators=[
        ('lgbm', lgbm),
        ('catboost', catboost),
        ('xgboost', xgboost),
#         ('knn', knn),
        ('hist_gbc', hist_gbc)
    ],
    voting='soft'  
)

# Fit the model
voting_model.fit(Xtrain, ytrain)

# Make predictions
y_pred = voting_model.predict(Xtest)
```

```python
y_pred = encoder.inverse_transform(y_pred)
```

```python
predictions = pd.DataFrame({'id': testid, 'Target': y_pred})
predictions.to_csv('submission.csv', index=False)
```
